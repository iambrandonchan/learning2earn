```python
import pandas as pd
import requests
import json
```

## Cities

```python
r = requests.get('http://api.parksr.us/cities?page=1')
```

```python
ex = json.loads(r.text)
```

```python
cities = { 'country': [],
           'id': [],
           'latitude': [],
           'longitude': [],
           'name': [],
           'num_parks': [],
           'state': []
         }
```

```python
ex['objects'][0].keys()
```

```python
page = 1
```

```python
base_url = 'http://api.parksr.us/cities?page='
```

```python
max_ = None

while max_ is None or page <= max_:
    url = base_url + str(page)
    r = requests.get(url)
    result = json.loads(r.text)
    if max_ is None:
        max_ = result['total_pages']
    objects = result['objects']
    for o in objects:
        for key in cities:
            cities[key].append(o[key])
    page += 1
```

### Pandas magic

```python
df = pd.DataFrame(cities)
```

```python
df.info()
```

```python
df['latitude'] = df['latitude'].astype('float32')
df['longitude'] = df['longitude'].astype('float32')
```
